```python
# def get_all_link(url_list,headers):      
for url in url_list:   
    resp = requests.get(url,headers = headers)
    resp.encoding = "UTF-8"
    html = etree.HTML(resp.text)

    divs = html.xpath("/html/body/div[5]/div[1]/div[4]/div")
    for div in divs:
        user = div.xpath("./div/a/div/div[1]/span[1]/text()")[0]
        link = "https://dianping.yiche.com" + div.xpath("./a/@href")[0]
        url_review_list.append(link)

        review_date = div.xpath("./div/a/div/div[1]/p/span/text()")[0]
        model = div.xpath("./a/div[1]/div/p/text()")[0]
        price = div.xpath("./a/div[1]/div/div/span[1]/span[2]/text()")[0]
        purchase_date = div.xpath("./a/div[1]/div/div/span[3]/span[2]/text()")[0]
        hot_review = div.xpath("./div/a/div/div[2]/div/img/@alt")

        if hot_review == []:
            hot_review = " "
        else:
            hot_review = hot_review[0]


        star_1 = div.xpath("./a/div[1]/div/div/div/div[1]/@class")[0]
        star_2 = div.xpath("./a/div[1]/div/div/div/div[2]/@class")[0]
        star_3 = div.xpath("./a/div[1]/div/div/div/div[3]/@class")[0]
        star_4 = div.xpath("./a/div[1]/div/div/div/div[4]/@class")[0]
        star_5 = div.xpath("./a/div[1]/div/div/div/div[5]/@class")[0]
        star_list = [star_1,star_2,star_3,star_4,star_5]
        rating = 0
        for j in star_list:
            if j == "star-gray active":
                rating += 1
            else:
                pass
    #     print(user,link,rating,review_date,model,price,purchase_date,hot_review)

    #         with open("长安1.csv", "a", encoding="utf-8") as f:
    #                 f.write("{},{},{},{},{},{},{},{}".format(user,link,rating,review_date,model,price,purchase_date,hot_review))
    #                 f.write("\n")
        comment_content = div.xpath("./a/div[2]/div[1]//text()")

        result1["user"].append(user)
        result1["link"].append(link)
        result1["rating"].append(rating)
        result1["review_date"].append(review_date)
        result1["model"].append(model)
        result1["price"].append(price)
        result1["purchase_date"].append(purchase_date)
        result1["hot_review"].append(hot_review)
        result1["comment_content"].append(comment_content)

        time.sleep(1)


    df = pd.DataFrame(result1)
    df.to_csv("ModelY1.csv", encoding="utf_8_sig")


    print("这一页完成！")
    print(url)
```

```python
def get_comment(url):
    r = requests.get(url,verify=False)
    r.encoding = "utf-8"
    comment = etree.HTML(r.text)

#     user = comment.xpath("/html/body/div[4]/div[1]/div[1]/div[3]/div[2]/text()")[0]
#     print(user)

    space = comment.xpath("/html/body/div[4]/div[1]/div[1]/div[6]/div[3]/div/div[1]/p[1]/span/text()")
    outlook = comment.xpath("/html/body/div[4]/div[1]/div[1]/div[6]/div[3]/div/div[2]/p[1]/span/text()")
    interior = comment.xpath("/html/body/div[4]/div[1]/div[1]/div[6]/div[3]/div/div[3]/p[1]/span/text()")
    power = comment.xpath("/html/body/div[4]/div[1]/div[1]/div[6]/div[3]/div/div[4]/p[1]/span/text()")
    manipulation = comment.xpath("/html/body/div[4]/div[1]/div[1]/div[6]/div[3]/div/div[5]/p[1]/span/text()")
    comfortable = comment.xpath("/html/body/div[4]/div[1]/div[1]/div[6]/div[3]/div/div[6]/p[1]/span/text()")
    continuity = comment.xpath("/html/body/div[4]/div[1]/div[1]/div[6]/div[3]/div/div[7]/p[1]/span/text()")
    purchase_place = comment.xpath("/html/body/div[4]/div[1]/div[1]/div[5]/div[3]/div[2]/div//text()")[0]
    
#     rating_list = [space,outlook,interior,power,manipulation,comfortable,continuity]
#     for i in range(7):
#         if rating_list[i] == []:
#             rating_list[i] == "NA"
    
#     comment_content = comment.xpath("/html/body/div[6]/div[1]/div[1]/div[6]/div[4]/div[1]//text()")
#     print(user, space,outlook,interior,power,manipulation,comfortable,continuity,comment_content)

#     with open("五菱宏光3.csv", "a", encoding="utf-8") as h:
#         h.write("{},{},{},{},{},{},{},{},{}".format(user,rating_list[0],rating_list[1],rating_list[2],
#                                                    rating_list[3],rating_list[4],rating_list[5],
#                                                    rating_list[6],comment_content))

#         h.write("{}".format(comment_content))
#         h.write("\n")


    result2["space"].append(space)
    result2["outlook"].append(outlook)
    result2["interior"].append(interior)
    result2["power"].append(power)
    result2["manipulation"].append(manipulation)
    result2["comfortable"].append(comfortable)
    result2["continuity"].append(continuity)
    result2["purchase_place"].append(purchase_place)
#     result2["comment_content"].append(comment_content)

    df = pd.DataFrame(result2)
    df.to_csv("ModelY2.csv", encoding="utf_8_sig")
    time.sleep(0.5)
    print("成功")
    
```
